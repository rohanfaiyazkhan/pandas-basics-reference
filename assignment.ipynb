## Data Manipulation

### Selecting Columns

We can access an entire column by directly using its column label as an object member. This returns the column as a series.

```python
df.bathrooms
```

```python
type(df.bathrooms)
```

We can also run `describe` on this series

```python
df.bathrooms.describe()
```

We can also index multiple columns by name. Doing so returns another dataframe with the selected columns

```python
df[['bathrooms', 'bedrooms']]
```

And since it returns another dataframe, we can also call describe with this.

```python
df[['bathrooms', 'bedrooms', 'beds', 'guests_included']].describe()
```
