## Data Manipulation

### Selecting Columns

We can access an entire column by directly using its column label as an object member. This returns the column as a series.

```python
df.bathrooms
```

```python
type(df.bathrooms)
```

We can also run `describe` on this series

```python
df.bathrooms.describe()
```

We can also index multiple columns by name. Doing so returns another dataframe with the selected columns

```python
df[['bathrooms', 'bedrooms']]
```

And since it returns another dataframe, we can also call describe with this.

```python
df[['bathrooms', 'bedrooms', 'beds', 'guests_included']].describe()
```

### Row Selection by Index

We can select rows by index using `df.iloc` which returns a series with the row and column labels.

```python
df.iloc[1] # second row
```

```python
df.iloc[2] # third row
```

```python
type(df.iloc[1])
```

We can also pass range or series as the index to `df.iloc`. Doing this will return a new dataframe.

```python
df.iloc[2:10] # select 3rd to 9th rows
```

```python
df.iloc[[1,3,4,9,19,45]] # select the 2nd, 4th, 5th, 10th, 20th and 46th rows
```

And ofcourse we can describe this new dataframe.

```python
df.iloc[[1,3,4,9,19,45]].describe()
```

### Selecting rows based on condition

`df.loc` is more generalized than `iloc` and while it is not as optimized (in terms of memory usage), we can do a lot more like indexing based on conditions.

```python
df.loc[df['beds'] > 3]
```

We can intuit that here we are indexing based on the condition of which AirBNB listings have more than 3 beds.

```python
df.loc[(df['bedrooms'] > 3) & (df['beds'] > 3)] 
```

Like with other conditions, we can combine logical conditions. However it is important that we use `&` and `|` instead of `and` and `or`. This is because the former have special uses in Pandas and they refer to elementwise logical operation. 

```python
df.loc[df['bedrooms'] > df['beds']]
```

Here we see that we can compare one column to another when indexing.

```python
df.loc[df['room_type'] == 'Private room'].describe()
```

```python
df.loc[df['room_type'].isin(['Hotel room', 'Private room'])].describe()
```

### Every time we select the rows or part of the tables, actually we are creating a new variable

```python
df_beds = df.loc[df['beds'] > 3]
df_beds, df
```

We can see here that these two are different
